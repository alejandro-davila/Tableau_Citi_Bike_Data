```python
import pandas as pd
```

```python
files = [
        'CitiBike_Data/202307-citibike-tripdata.csv',
        
        'CitiBike_Data/202308-citibike-tripdata.csv',
        
        'CitiBike_Data/202309-citibike-tripdata.csv']
```

```python
citi_bike_df = pd.DataFrame()
```

```python
for file in files:
    data = pd.read_csv(file)
    citi_bike_df = pd.concat([citi_bike_df, data], axis=0, ignore_index=True)
    
```
